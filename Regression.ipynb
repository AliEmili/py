```python
from sklearn.linear_model import LinearRegression
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(1,10)
y = np.array([20,25,26,31,29,30,35,36,46])
```

```python
plt.scatter(x,y) # sakhtane noghate 
plt.show()
```

```python
x = x.reshape(-1, 1) # taranehade kardan bordar, sootoonish mikone
y = y.reshape(-1, 1)
reg = LinearRegression()
reg.fit(x,y) # dade haye test ro midim behesh (train)
```

```python
pred = reg.predict(x) 
plt.scatter(x,y) 
plt.plot(x,pred)
plt.show()
```
